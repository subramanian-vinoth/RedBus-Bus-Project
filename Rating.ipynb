```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains
import pandas as pd
import time
```

```python
# Initialize WebDriver
driver = webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.redbus.in/bus-tickets/bangalore-to-tirupathi?fromCityId=122&toCityId=71756&fromCityName=Bangalore&toCityName=Tirupati&busType=Any&onward=21-Jan-2025")
time.sleep(5)

# Wait for the button to load and click
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@class='button']")))
button = driver.find_element(By.XPATH, "//div[@class='button']")
button.click()

wait = WebDriverWait(driver, 10)
# scrolling = True
# while scrolling:
#     old_page_source = driver.page_source
#     actions = ActionChains(driver)    
#     actions.send_keys(Keys.PAGE_DOWN).perform()
#     time.sleep(6)
    
#     new_page_source = driver.page_source   # new page after scrolling

#     if old_page_source == new_page_source:
#         scrolling = False


scrolling = True
old_page_source = driver.page_source
max_attempts = 5  # Maximum attempts to detect no new content
attempts = 0

while scrolling:
    # Use Keys.PAGE_DOWN to scroll
    body = driver.find_element(By.TAG_NAME, "body")
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(3)  # Pause to allow content to load

    # Capture new page source
    new_page_source = driver.page_source

    # Check if the page source has changed
    if old_page_source == new_page_source:
        attempts += 1
        if attempts >= max_attempts:  # Stop after max attempts
            scrolling = False
    else:
        attempts = 0  # Reset attempts if new content is loaded

    old_page_source = new_page_source

bus_name = driver.find_elements(By.CSS_SELECTOR, "div[class='travels lh-24 f-bold d-color']")
bus_type = driver.find_elements(By.CSS_SELECTOR, "div[class='bus-type f-12 m-top-16 l-color evBus']")
Departing_time = driver.find_elements(By.XPATH, "//div[@class='dp-time f-19 d-color f-bold']")
Reaching_time = driver.find_elements(By.CSS_SELECTOR, "div[class='bp-time f-19 d-color disp-Inline']")
Duration = driver.find_elements(By.CSS_SELECTOR, "div[class='dur l-color lh-24']")
Price = driver.find_elements(By.XPATH, '//div[@class="fare d-block"]//span')
rating = driver.find_elements(By.XPATH, "//div[@class='column-six p-right-10 w-10 fl']")
seat_availability = driver.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")

Kerela_bus_name = []
for name in bus_name:
    Kerela_bus_name .append(name.text)

Kerela_bus_type = []
for b_type in bus_type:
    Kerela_bus_type .append(b_type.text)

Kerela_Departing  = []
for Departing in Departing_time:
    Kerela_Departing.append(Departing.text)

Kerela_Reaching_time = []
for r in Reaching_time:
    Kerela_Reaching_time.append(r.text)

Kerela_Duration  = []
for d in Duration:
    Kerela_Duration.append(d.text)

Kerela_Price  = []
for p in Price:
   Kerela_Price.append(p.text)

Kerela_seat_availability  = []
for s in seat_availability:
   Kerela_seat_availability.append(s.text)


Kerela_rating = []
for element in rating:
    # Get the text of the element
        rating_text = element.text.strip()

        if rating_text:  # Check if the element contains text
            # Split the text to extract the rating (assuming the first part is the rating)
            try:
                Kerela_rating.append(float(rating_text.split("\n")[0]))
            except ValueError:  # Handle unexpected formats gracefully
                Kerela_rating.append(0.0)
        else:  # If the element is empty, assign 0
            Kerela_rating.append(0.0)

print("Completed")
```

```python
kerela_bus_data = {
            'busname': Kerela_bus_name,
            'bustype': Kerela_bus_type,
            'Departing_time': Kerela_Departing,
            'Reaching_time': Kerela_Reaching_time,
            'Duration': Kerela_Duration,
            'Price': Kerela_Price ,
            'rating': Kerela_rating ,
            'seats_available': Kerela_seat_availability
        }
```

```python
b = pd.DataFrame(kerela_bus_data)
b
```
